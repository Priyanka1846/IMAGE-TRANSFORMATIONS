```python
# Import

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('nature.jpg')
```

```python
img.shape
```

```python
plt.imshow(img[:,:,::-1])
plt.title('Original image')
plt.show()
```

```python
#Image Translation

tx,ty=200,400
trans=np.float32([[1,0,tx],[0,1,ty]])
trans_img=cv2.warpAffine(img,trans,(1536,1024))
```

```python
plt.imshow(trans_img[:,:,::-1])
plt.title('Translated Image')
plt.axis('on')
plt.show()
```

```python
# Image Scaling 

fx,fy=3.0,1.0
scaled=cv2.resize(img,None,fx=fx,fy=fy, interpolation=cv2.INTER_LINEAR)
```

```python
plt.imshow(scaled[:,:,::-1])
plt.title('Scaled Image')
plt.axis('on')
plt.show()
```

```python
# Image Shearing

shear_matrix=np.float32([[1,0.3,0],[0.7,1,0]])
shaer_img=cv2.warpAffine(img,shear_matrix,(1536,1024))
```

```python
plt.imshow(shaer_img[:,:,::-1])
plt.title("Sheared Image")
plt.axis('on')
plt.show()
```

```python
# Image Reflection 

reflected=cv2.flip(img,1)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,2)
plt.imshow(reflected[:,:,::-1])
plt.title('Reflected Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Image Rotation

(height,width)=img.shape[:2]
angle=45
center=(width//2,height//2)
rot=cv2.getRotationMatrix2D(center,angle,1)
rot_img=cv2.warpAffine(img,rot,(width,height))
```

```python
plt.imshow(cv2.cvtColor(rot_img, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.axis('off')
```

```python
# Image Cropping

x,y,w,h=50,100,300,200
# cropped = img[100:300, 50:350]
cropped=img[y:y+h,x:x+w]
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.imshow(cropped[:, :, ::-1])
print("Shape of cropped image:", cropped.shape)
plt.title("Cropped Image")
plt.axis('on')

plt.tight_layout()
plt.show()
```
